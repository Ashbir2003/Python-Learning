# Dictionary : a dictionary is a built-in data type used to store data in key–value pairs.


Dictionary is a mutable, ordered, dynamic collection that stores data in key-value pairs. Keys must be unique and immutable, while values can be duplicate and of any data type.

> 🔑 What does that mean?
- Each key is like a label (it must be unique).
- Each value is the data associated with that key.
- You use the key to quickly access its value.

# General Syntax
dictionary_name = {
    key1: value1,
    
    key2: value2,
    
    key3: value3
}

# removal : In Python, there are 4 common ways to remove items from a dictionary.

1) Pop() : Removes the item with the specified key and returns its value.

Syntax

dictionary.pop(key)

```python
student_info = {
    'name' : "Asha",
    'age' : 20,
    'ispass' : True
}
removed_item = student_info.pop('age','ispass') 
print(student_info) 
print(removed_item)   # returns the value
```

# 2. popitem()

Removes and returns the last inserted key-value pair.

returns the removed value along with key (ex: ('city', 'bangalore')

returns tuple(key -value)

```python
student_info = {
    'name': "Asha",
    'age': 20,
    'city': "bangalore"
}
removed_item = student_info.popitem()
print(removed_item)   # returns tuple(key -value)
print(student_info)
```

# Key() : The keys() method returns all the keys present in a dictionary.

- ✅ keys() returns all dictionary keys.

- ✅ It returns a dict_keys object, not a list.

- ✅ The returned object is dynamic—if the dictionary changes, the keys view updates automatically.

```python
student_info = {
    'name': "Asha",
    'age': 20,
    'city': "bangalore"
}
key = student_info.keys()
key
```

#### converting keys to list

```python
print(list(key))
```

```python
# Using keys() in a Loop
student_info = {
    'name': "Asha",
    'age': 20,
    'city': "bangalore"
}
for key in student_info:
    print(key)
```

# value() : The values() method returns all the values stored in a dictionary.

```python
student_info = {
    'name': "Asha",
    'age': 20,
    'city': "bangalore"
}
value = student_info.values()
print(value)
print(list(value))
```

# items() : The items() method returns iterable view of both keys and values as pairs (tuples).

```python
student_info = {
    'name': "Asha",
    'age': 20,
    'city': "bangalore"
}
item = student_info.items()
print(item)
print(list(item))
```

## sorting dictionary by keys: u can use sorted function along with the items() method to sort the dictionary item (key-value pair) and then construct new dictionary from sorteditem

## Dictionary Comprehension in Python

Dictionary comprehension is a short and concise way to create dictionaries in a single line of code.

It works similarly to list comprehension.

##### Syntax
{expression_for_key: expression_for_value for item in iterable}

```python
# Creating dictionary comprehension using number
square_dict = {x:x**2 for x in range(1,10)}
print(square_dict)
```

### creating dict comprehension to create a dictionary mapping words to their lengths

```python
words = ['apple', 'banana', 'mango']
length_words = {word : len(word) for word in words}
print(length_words)
```

# Features of Set in Python
##Unordered
- Elements do not have a fixed position.
- You cannot access items using an index.

```python
colors = {'red', 'yellow', 'blue'}
colors
```

### Unique Elements
- Duplicate values are automatically removed.

```python
colors = {'red', 'yellow', 'blue', 'red'}
colors
```

### Mutable
- You can add or remove elements after creation.

```python
colors.add('papaya')
colors.remove('red')
colors
```

### Dynamic Size
- A set can grow or shrink as elements are added or removed.
### Fast Membership Testing
- Checking whether an element exists is very efficient.

```python
'papaya' in colors
```

### Elements Must Be Immutable
- Items like integers, strings, tuples can be stored.
- Lists and dictionaries cannot be stored directly in set.

```python
s = {1, "hello", (1, 2)}
s
# s = {1, "hello", [1, 2]}   : error
```

### No Indexing or Slicing

Since sets are unordered, indexing is not allowed.

```python
# s[0]  ❌ Error
```

# Manipulating sets 

## set operations refers to various mathematicals operation
## Union(|) : Combines elements from tw sets, resulting new set, combining uniques values from both sets

```python
set1 = {1,2,3,4,5}
set2 = {5,6,7,8,9,10}
set3 = set1 | set2
set3
```

## Intersection : gives common element from both sets

```python
set1 = {1,2,3,4,5}
set2 = {5,6,7,8,9,10}
set3 = set1 & set2
set3
```

## Difference () - perform minus

```python
set1 = {1,2,3}
set2 = {1,4}
set3 = set1 - set2
set3
```

```python
### Q: given two sets, find items that appear in both-ignore duplicates
l1 = [1,2,2,3,4]
l2 = [3,3,4,5]
l1_to_set = set(l1)
l2_to_set = set(l2)
print(l1_to_set)
print(l2_to_set)
intersect = l1_to_set  & l2_to_set
print(intersect)
```

## check if given two strings are anagrams or not

# Intersection and differences
## dictionary(keys)

```python
dict1 = {'a':1 , 'b':2, 'c':3}
dict2 = {'a':1 , 'd':2, 'c':3}
common_keys = dict1.keys() | dict2.keys()
combined_keys = dict1.keys() & dict2.keys()
unique_keys = dict1.keys() - dict2.keys()
print(common_keys)
print(combined_keys)
print(unique_keys)
```

```python
# create a dictionary of word freq ignoring common stopwords
text = 'python is great and python is fast but python is also simple'
stopwords = {"is", "and", "but"}
new_word = text.split()
print(new_word)
nn = {word: new_word.count(word) for word in new_word if word not in stopwords}
print(nn)
```

```python
text = 'python is great and python is fast but python is also simple'
stopwords = {"is", "and", "but"}

# 1. Split the text into individual words
# 2. Filter out words that are in the stopwords set
# 3. Join the filtered words back into a single string
filtered_text = " ".join({word: text.count(word) for word in text.split() if word not in stopwords})

print(filtered_text)
# Output: 'python great python fast python also simple'
```
