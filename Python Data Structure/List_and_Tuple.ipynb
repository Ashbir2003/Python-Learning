# What is a List in Python?

A list is a built-in data type , enclosed in []

used to store multiple items in a single variable.

Lists are ordered, changeable (mutable), and allow duplicate values.

# Characteristics of Lists
- 1. Ordered

Items are stored in the order they are added.
- 2. Mutable (Can be Changed)

You can modify, add or remove list elements after creation.
- 3. Allows Duplicate Values
- 4. Can Store Different Data Types
- 5. Supports Indexing
    - Each element has an index number starting from 0.
- 6 support Slicing and nested list

## 2. insert() : Adds an element at a specific position.
"The insert() method is used to add an element at a specific index in a list. The existing elements are shifted to the next positions, and the original list is modified."

Syntax
- list.insert(index, element)
- index → Position where the element should be inserted.
- element → Value to be added.

```python
number = [10, 30, 43]
a = number.insert(1,3)
print(number)
print(a)  # returns none because it modifies original one
```

### Important Points

- ✅ Inserts element at a specified index.

- ✅ Existing elements shift to the right.

- ✅ Modifies the original list.

- ✅ Returns None.

- ❌ Does not replace existing elements.

- Return value= none

# copy() Method in Python List

The copy() method is used to create a copy of a list.

This is useful when you want another list with the same elements, but changes to one list should not affect the other.

apply only list   not nested list

```python
# without copy
a = [1,2]
b = a
b[0] = 3
print(a)
print(b)
```

```python
# with copy
a = [1,2]
b = a.copy()
b[0] = 3
print(a)
print(b)
```

```python
a = [2,3,4]
b = a.copy()
print(a)
print(b)
b.append(44)
print(b)
print(a)
```

```python
# copy wont work for nested list
a = [[1,2],[2,3]]
b = a.copy()
b[0][0] = 100
print(a)
print(b)
```

```python
# do to so use
import copy 
a = [[1,2],[2,3]]
b = copy.deepcopy(a)
b[0][0] = 100
print(a)
print(b)
```

# Filtering and transformation with nested list comprehensions:

```python
lst = []
number = [[1,2,3],[3,4,5],[6,7,8]]
for row in number:
    print(row)
    for item in row:
        print(item)
        if item % 2 == 0:
            lst.append(item**2)
lst
```

```python
number = [[1,2,3],[3,4,5],[6,7,8]]
lst = [item**2 for row in number for item in row if item % 2 == 0]
lst
```

```python
nested_matrix = [ [ 1,2,3],[4,5,6],[7,8,9]]
lst = [i for row in nested_matrix for i in row if i >3 ]
lst
```

## converting nested loop into single loop by list compression

```python
# without list comprehension
nested_matrix = [ [ 1,2,3],[4,5,6],[7,8,9]]
flattened = []
for row in nested_matrix:
    print(row)
    for x in row:
        print(x)
        flattened.append(x)
flattened
```

```python
# with list comprehension
nested_matrix = [ [ 1,2,3],[4,5,6],[7,8,9]]
lst = [x for row in nested_matrix for x in row]
lst
```

```python
# converting single list into nested
arr_2d = []
start = 0
for incr in range(1,4):
    print(incr)
    arr_2d.append(flattened[start:incr*3])
    print(start)
    start = incr*3
    print("after", start)
arr_2d
```
