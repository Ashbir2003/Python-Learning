# String Slicing in Python

### String slicing means taking a part (slice) of a string using index.

### We use slicing to get multiple characters from a string.
#### In Python, every character in a string has a position called an index.

Syntax
- string[start : stop : step]
- start → where slicing starts
- stop → where slicing ends (not included)
- step → how many positions to move
- Character	P	y	t	h	o	n
- Positive Index	0	1	2	3	4	5
- Negative Index	-6	-5	-4	-3	-2	-1

```python
# define the variable text
text = "Python"

#slice the string using from index 1 to 4(4 is not included) 
sliced_string = text[1:4]
sliced_string1 = text[1:4:2]
sliced_string2 = text[1:-4]
sliced_string3 = text[-2:4]
sliced_string4 = text[1::2]
print(sliced_string)
print(sliced_string1)
print(sliced_string2)
print(sliced_string3)
print(sliced_string4)
```

- Important:

- Start index is included
- Stop index is excluded
