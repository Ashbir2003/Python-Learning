# Pandas : Pandas is a popular Python library used for data analysis and data manipulation. It provides powerful data structures that make it easy to work with tables, spreadsheets, CSV files, databases, and other structured data.

### Key Features
- Read and write data from files (CSV, Excel, JSON, SQL, etc.)
- Clean and transform data
- Filter, sort, and group data
- Handle missing values
- Perform statistical analysis
- Work with time-series data

## Main Data Structures

### 1. Series
- One-dimensional (only one column)
- Stores values of the same or mixed data types
- Has an index and values

General syntax:

pd.Series(data)

### 2. DataFrame
A two-dimensional table with rows and columns (similar to an Excel spreadsheet).

```python
# creating a series from numpy array
numpy_array = np.array([1,2,3,4,5])
print(numpy_array)
series_from_nparray = pd.Series(numpy_array)
print(series_from_nparray)
```

```python
# creating series with custom index
data = [10,20,30,40,50]
custom_index = pd.Series(data, index=['A', 'B', 'C','D','E'])
print(custom_index)             
```

## slicing of series


```python
data = [10,20,30,40,50]
series_Data = pd.Series(data)
sliced_series = series_Data[1:3]
print(sliced_series)
```

### Series from dictionary

```python
dict = {"asha":85, "varsha":90}
series = pd.Series(dict)
print(series)
print(series.index)               # Check the Index
print(series.values)             # Check Values
print(series.dtype)             # Check the data type
```

# Dataframe: 
A DataFrame is the most important data structure in Pandas.

It is a 2-dimensional table made up of rows and columns, similar to:

- An Excel sheet
- A SQL table
- A CSV file

```python
# creaing a dataframe from dictionary
dict = {"Name": ["Asha", "Usha", "Nisha"],
       "Age": [22,23,24]}
df_from_dict = pd.DataFrame(dict)
print("dataframe created from dictionary")
print(df_from_dict)
```

## creating a DataFrame from list of list with custom column names

```python
data_list = [["Asha",24], ["Usha", 25], ["Nisha", 26]]
list_to_df = pd.DataFrame(data_list,columns = ["name", "age"])
print(list_to_df)
```

## creating dataframe from csv file

# Indexing a Pandas DataFrame

dataframes allow varoius methods of indexing, including selecting columns, filtering rows and more

### selecting single column

```python
df_from_dict
```

```python
# selecting single column
single_column = df_from_dict['Name']
print(single_column)
```

## selecting multiple columns

```python
multi_column = df_from_dict[['Name', 'Age']]
print(multi_column)
```

## filtering rows based on condition

```python
young_people = df_from_dict[df_from_dict['Age'] <24]
print(young_people)
```
